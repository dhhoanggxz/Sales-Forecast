# 1. Tổng quan về đề tài

## Đề tài: Sales Forecasting for a Store

#### Sales Forecast là gì?
Sale forecast là dự báo bán hàng, chỉ sự ước tính về các chỉ số bán hàng trong tương lai. Các chỉ số bán hàng đó được đo lường bằng doanh thu hoặc số lượng sản phẩm. Và thời gian dự báo có thể ngắn hạn theo tuần, theo tháng, theo từng quý hoặc dài hạn hơn là theo năm. 

#### Tầm quan trọng của Sales Forecast: 
- Thứ nhất, thông qua Sales Forecast, các doanh nghiệp có thể xác định khối lượng sản phẩm cần phải sản xuất hình thành trong tương lai như thế nào dựa trên những nguyên liệu có sẵn. Chẳng hạn như những trang thiết bị, máy móc, phương tiện, nguồn vốn, không gian sản xuất, nhân lực…

- Thứ hai, các doanh nghiệp có thể ước tính và ước lượng được tổng doanh thu mà doanh nghiệp sẽ có được trong một khoảng thời gian nhất định từ hoạt động bán hàng. Điều này trở thành một chỉ tiêu, một cam kết để thúc đẩy những bộ phận bán hàng phải chịu trách nhiệm bảo đảm đạt được mục tiêu trong khoảng thời gian đã được dự báo từ trước đó.

- Thứ ba, các doanh nghiệp sẽ nhìn thấy được sản phẩm của doanh nghiệp hình thành trong tương lai cũng như doanh thu mà doanh nghiệp sẽ nhận được. Từ đó, doanh nghiệp sẽ có các chiến lược phát triển mới, xây dựng, mở rộng cơ sở sản xuất, thay đổi cơ cấu và thúc đẩy quá trình sản xuất để đáp ứng được nhu cầu của việc bán ra hàng hóa hay chuyển nguồn lực để sản xuất sang những sản phẩm khác.

#### Dự báo với mô hình Random Forest Regression
- Mô hình Random Forest Regression là một kỹ thuật học máy được sử dụng để dự đoán giá trị số (numerical values). Nó thuộc vào họ các mô hình gọi là "Ensemble Learning," trong đó, nhiều mô hình được kết hợp để tạo ra một mô hình mạnh mẽ hơn và ổn định hơn so với việc sử dụng một mô hình đơn lẻ.
- Ensemble Learning là một kỹ thuật máy học kết hợp các dự đoán từ nhiều mô hình để tạo ra một dự đoán chính xác và ổn định hơn. Đây là một phương pháp sử dụng sự thông minh tổng hợp từ nhiều mô hình để nâng cao hiệu suất tổng thể của hệ thống.
- Random Forest sử dụng nhiều cây quyết định (Decision Trees) để tạo ra một "rừng" (forest) các cây quyết định. Mỗi cây quyết định có phương sai cao, nhưng khi chúng ta kết hợp tất cả chúng song song, phương sai cuối cùng sẽ thấp vì mỗi cây quyết định được đào tạo hoàn hảo trên dữ liệu mẫu cụ thể, và do đó, kết quả không phụ thuộc vào một cây quyết định mà là trên nhiều cây quyết định. Trong trường hợp phân loại, kết quả cuối cùng được xác định bằng cách sử dụng bộ phân loại thông qua việc bỏ phiếu đa số. Trong trường hợp hồi quy, kết quả cuối cùng là giá trị trung bình của tất cả các kết quả. Phần này được gọi là Tổng hợp.
- Khi cần dự đoán một giá trị mới, mỗi cây trong rừng đưa ra dự đoán của mình. Kết quả dự đoán cuối cùng là trung bình (đối với Regression) hoặc là giá trị phổ biến (đối với Classification) của các dự đoán từ tất cả các cây.
- Ngoài ra, Random Forest giúp giảm overfit bằng cách kết hợp dự đoán từ nhiều cây. Từ đó sẽ đưa ra độ chính xác cao hơn so với việc sử dụng một cây quyết định đơn lẻ.

## Vì sao không loại bỏ các giá trị outlier trong bộ dữ liệu trên?
- Trong bài báo cáo này sẽ xây dựng mô hình Random Forest Regressor, bản chất của mô hình này sẽ không bị nhạy cảm với các outlier như các mô hình hồi quy thông thường (do nó sử dụng nhiều decision trees)([Phần 2.2](https://www.listendata.com/2014/11/random-forest-with-r.html)). Sự kết hợp từ nhiều cây quyết định giúp giảm ảnh hưởng của các outlier đến quyết định cuối cùng. Ngoài ra, mô hình này có tính phi tham số. Do đó, nó không phụ thuộc vào giả định về phân phối của dữ liệu và điều này giúp cho mô hình ít bị ảnh hưởng hơn bởi các outlier.
- Các outlier trong bộ dữ liệu trên là doanh số trung bình những tuần có ngày lễ của các cửa hàng, những giá trị được xem là các outlier này giúp cho ta thấy được sự khác biệt về doanh số giữa các tuần bình thường với các tuần có ngày lễ.
- Dữ liệu của chúng ta áp dụng vào mô hình được lựa chọn một cách không ngẫu nhiên và các điểm outlier này chiếm một phần lớn trong dữ liệu, loại bỏ chúng có thể dẫn đến một mô hình biểu diễn không đúng  cho toàn bộ dữ liệu.
- Một số mức phân vị thường được sử dụng để loại bỏ outliers: 
    - [5% và 95%](https://github.com/TienDung0208/Image/blob/main/5_95.jpg)
    - [10% và 90%](https://github.com/TienDung0208/Image/blob/main/10_90.jpg)
    - [1% và 99%](https://github.com/TienDung0208/Image/blob/main/1_99.jpg)
    
- Đoạn code để biểu diễn tần suất của các dữ liệu trung tâm trong những biểu đồ trên:
```python
# lần lượt thay quantile = [0.05, 0.95]; [0.1, 0.9]; [0.01, 0.99]
low, high = df['Weekly_Sales'].quantile([0.05, 0.95]) 
mask_weekly_sales = df['Weekly_Sales'].between(low, high)
temp = temp[mask_weekly_sales]
plt.hist(temp['Weekly_Sales']);
```
    
&Rightarrow; Từ kết quả của các mức phân vị trên, ta thấy rằng outliers bắt đầu xuất hiện khi sử dụng mức phân vị 1% và 99%.
- Đoạn code giữ lại 98% dữ liệu trung tâm, bỏ đi 1% ở mỗi đầu:
```python
quantile_99 = np.percentile(df['Weekly_Sales'], 99)
# Kết quả: quantile_99 = 106566.35
quantile_1 = np.percentile(df['Weeklu_Sales'], 1)
# Kết quả: quantile_1 = 7.44
# Đếm số điểm outliers của 1% dữ liệu loại bỏ ở mỗi đầu:
outliers_1 = 0
outliers_2 = 0
for sales in df['Weekly_Sales']:
    if sales > 106566:
        outliers_1 += 1
    elif sales < 7.44:
        outliers_2 += 1
''' 
Kết quả: outliers_1 = 4203
         outliers_2 = 4189 
'''
```
&Rightarrow; Kết quả này đã cho ta thấy rằng các điểm outlier chiếm 1 phần lớn trong dữ liệu.

# 6. Xây dựng mô hình Random Forest Regression

## Phương pháp WMAE

- Phương pháp WMAE (Weighted Mean Absolute Error) là một thước đo độ chính xác thường được sử dụng trong các bài toán dự đoán và đánh giá mô hình, đặc biệt là trong lĩnh vực dự đoán doanh số bán hàng hoặc doanh số hàng hóa.
- Công thức của WMAE được xây dựng bằng cách tính trung bình cộng của giá trị tuyệt đối của sự chênh lệch giữa giá dự đoán và giá trị thực, với mỗi giá trị được nhân với một trọng số tương ứng. Công thức chung có thể được biểu diễn như sau:

![WMAE](WMAE.jpg)

Trong đó:
- n là số hàng
- ŷ là doanh số dự đoán
- yi là doanh số bán hàng thực tế
- wi là trọng số. w = 5 nếu tuần là tuần nghỉ lễ, 1 nếu không là tuần

Với số liệu này, sai số ở các tuần nghỉ lễ có trọng số gấp 5 lần so với các tuần bình thường. Vì vậy, điều quan trọng hơn là phải dự đoán chính xác doanh số bán hàng vào các tuần nghỉ lễ.

# Tạo bộ dữ liệu Train - Test


Vì cột Date của ta chứa các giá trị liên tục, nghĩa là các giá trị ngày được sắp xếp theo thứ tự thời gian, chia ngẫu nhiên có thể làm mất đi tính chất thứ tự này. Trong nhiều tình huống, đặc biệt là trong phân tích dữ liệu chuỗi thời gian, việc duy trì thứ tự thời gian là rất quan trọng. Nên ta sẽ chia dữ liệu 1 cách thủ công dựa trên tỷ lệ 70% cho tập train và 30% cho tập test.

````python
train_data = df_new[:int(0.7*(len(df_new)))]
test_data = df_new[int(0.7*(len(df_new))):]

target = "Weekly_Sales"
features = [col for col in df_new.columns.to_list() if col not in [target]]

X_train = train_data[features]
X_test = test_data[features]
y_train = train_data[target]
y_test = test_data[target]
````

````python
# Giữ giá trị của tập train và test cùng 1 lúc
X = df_new[features]
````

````python
# Bỏ cột Date vì đã có đủ thông tin (tuần trong năm)
X_train = X_train.drop(['Date'], axis=1)
X_test = X_test.drop(['Date'], axis=1)
````

# Hàm định nghĩa độ đo

````python
def wmae_test(test, pred): # WMAE for test 
    weights = X_test['IsHoliday'].apply(lambda is_holiday:5 if is_holiday else 1)
    error = np.sum(weights * np.abs(test - pred), axis=0) / np.sum(weights)
    return error
````

# Random Forest Regressor


````python
rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1, max_depth=35,
                           max_features = 'sqrt',min_samples_split = 10)

# Chuẩn hóa dữ liệu, giảm ảnh hưởng các giá trị ngoại lai
scaler = RobustScaler()

pipe = make_pipeline(scaler,rf)

# Dùng dữ liệu để huấn luyện
pipe.fit(X_train, y_train)

# Dự đoán trên tập huấn luyện
y_pred = pipe.predict(X_train)

# Dự đoán trên tập test
y_pred_test = pipe.predict(X_test)
````

````python
wmae_test(y_test, y_pred_test)
````
